# Extract and Parse Reddit Comments  
  
  This workbook will extract the previous days comments and this morning's comments.
  It then cleans the comments and extracts the Tickers mentioned.
  It will then perform sentiment analysis and summarize the day's results.
  It will update the previous file list with the new record.

## Environment Setup  
### Specify current comments to extract

```python
TOMORROW_SUBMISSION_ID = 'n9eiyu'
TOMORROW = '05_11_2021'
```

```python
import numpy as np
import pandas as pd
from datetime import datetime
import datetime, time
import praw as pr
import pickle
from wordcloud import WordCloud
import matplotlib.pyplot as plt
pd.set_option('display.max_colwidth', None)
import sys
sys.path.append('../Packages/')
import diamond_hands as dh
```

```python
raw_comment_dir = '../../Data/RawComments/'
processed_inputs_dir = '../../Data/ProcessedInputs/'
processed_comments_dir = '../../Data/Outputs/ProcessedComments/Tomorrow/'
wordcloud_dir = '../../Data/Outputs/Wordcloud/Tomorrow/'

today_comments_all = processed_inputs_dir + 'Today_All_Comment_Summary.bin'
today_comments_tickr = processed_inputs_dir + 'Today_Tickr_Comment_Summary.bin'
tomorrow_comments_all = processed_inputs_dir + 'Tomorrow_All_Comment_Summary.bin'
tomorrow_comments_tickr = processed_inputs_dir + 'Tomorrow_Tickr_Comment_Summary.bin'

logging_file = '../../Data/Log/log.bin'

REDDIT_CLIENT_ID = ""
REDDIT_CLIENT_SECRET = ""
REDDIT_USER_AGENT ="my user agent"
```

```python
# establish reddit connection
reddit = pr.Reddit(
    client_id=REDDIT_CLIENT_ID,
    client_secret=REDDIT_CLIENT_SECRET,
    user_agent=REDDIT_USER_AGENT,
)
```

## Comment Extraction  
Get comments using praw connection. Save results to file.  
**Note: this can take a while**

```python
tomorrow_submission = reddit.submission(id=TOMORROW_SUBMISSION_ID)
```

```python
start_time = datetime.datetime.now()
print('Started: %s' % start_time.strftime("%H:%M:%S.%f"))
print(tomorrow_submission.num_comments)
tomorrow_submission.comments.replace_more(limit=None)
diff = datetime.datetime.now() - start_time
print("Time taken: %s" % diff)
```

```python
tomorrow_comments = tomorrow_submission.comments.list()
tomorrow_file = raw_comment_dir+'Tomorrow/Tomorrow_'+TOMORROW+'.bin'
with open(tomorrow_file,'wb') as f:
    pickle.dump(tomorrow_comments,f)
```

## Comment Cleaning and Sentiment

```python
tomorrow_num_comments, tomorrow_summary_all, tomorrow_summary_tickr, processed_comments = dh.processComments(tomorrow_file)
```

```python
# Load historical data and append new records
with open(tomorrow_comments_all,'rb') as f:
    tomorrow_all_df = pickle.load(f)
    
with open(tomorrow_comments_tickr,'rb') as f:
    tomorrow_tickr_df = pickle.load(f)
    
tomorrow_summary_all = pd.Series(tomorrow_summary_all,index = tomorrow_all_df.columns)
tomorrow_all_df = tomorrow_all_df.append(tomorrow_summary_all,ignore_index=True)
tomorrow_summary_tickr2 = pd.DataFrame(tomorrow_summary_tickr,columns = tomorrow_tickr_df.columns)
tomorrow_tickr_df = tomorrow_tickr_df.append(tomorrow_summary_tickr2,ignore_index=True)
```

## Visualize Results

```python
dh.plotWordCloud(tomorrow_summary_tickr2,filename=wordcloud_dir+TOMORROW+'.png')
```

## Save Results

```python

with open(tomorrow_comments_all,'wb') as f:
    pickle.dump(tomorrow_all_df,f)
    
with open(tomorrow_comments_tickr,'wb') as f:
    pickle.dump(tomorrow_tickr_df,f)
```

```python
with open(processed_comments_dir+TOMORROW+'.bin','wb') as f:
    pickle.dump(processed_comments,f)
```

## Keep track of actions performed

```python
with open(logging_file,'rb') as f:
    model_log = pickle.load(f)

# Update comments processed
model_log['#Tomorrow_Processed'] += tomorrow_num_comments
model_log['Last_Tomorrow'] = TOMORROW

with open(logging_file,'wb') as f:
    pickle.dump(model_log,f)
```
