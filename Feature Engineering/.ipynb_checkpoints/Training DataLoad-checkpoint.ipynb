# Training DataLoader
This workbook will extract all of the data needed for training the model. It will also normalize data and save the scalers for future use.

## Environment Setups

```python
# Import packages

# General Utility
from datetime import datetime
import datetime, time
from datetime import timedelta

# Data processing
import numpy as np
import pandas as pd
import pickle
from sklearn.preprocessing import RobustScaler

# Diamond-Hands
import sys
sys.path.append('../Packages/')
import diamond_hands as dh
```

```python
# Stocks
stocks = ['SPY','AAPL','AMC','BB','F','GME','GRWG','MO','NIO','PLTR','RKT','SNDL','TLRY','TSLA','VIAC']

# Keys
ALPACA_API_KEY = ''
ALPACA_SECRET_KEY = ''
FRED_API_KEY =''

# Data Locations
processed_input_dir = '../../Data/ProcessedInputs/'
model_dir = '../../Data/Models/'

# Comment Files
today_all = processed_input_dir+'Today_All_Comment_Summary.bin'
today_tickr = processed_input_dir+'Today_Tickr_Comment_Summary.bin'
tomorrow_all = processed_input_dir+'Tomorrow_All_Comment_Summary.bin'
tomorrow_tickr = processed_input_dir+'Tomorrow_Tickr_Comment_Summary.bin'
```

## Dataloader for easily loading data

```python
dl = dh.MyDataLoader(ALPACA_API_KEY,ALPACA_SECRET_KEY,FRED_API_KEY,today_all,today_tickr,tomorrow_all,tomorrow_tickr)
```

## Load all the stock data and save to directory

```python
for stock in stocks:
    df = dl.loadData(stock, '2000-01-01', '2021-04-23')
    #df = df.replace([np.inf, -np.inf], 0)
    
    # SPlit out level data
    y = df[['Open']]
    df = df.drop(['Open','Change'],axis=1)
    col = df.pop("P_Change")
    df.insert(0, col.name, col)
    
    # Normalize data
    scaler_all = RobustScaler()
    scaler = RobustScaler()
    scaler.fit(df[:'2021-02-28'][['P_Change']])
    scaler_all.fit(df[:'2021-02-28'])
    df = pd.DataFrame(scaler_all.transform(df), columns=df.columns, index=df.index)
    
    # Save data
    with open(processed_input_dir+stock+'_output_data.bin','wb') as f:
        pickle.dump(y,f)

    with open(processed_input_dir+stock+'_input_data.bin','wb') as f:
        pickle.dump(df,f)

    with open(model_dir+stock+'_scaler_all.bin','wb') as f:
        pickle.dump(scaler_all,f)

    with open(model_dir+stock+'_scaler.bin','wb') as f:
        pickle.dump(scaler,f)
    
```
