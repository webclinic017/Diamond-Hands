# Extract and Parse Reddit Comments  
  
  This workbook will extract the previous days comments and this morning's comments.
  It then cleans the comments and extracts the Tickers mentioned.
  It will then perform sentiment analysis and summarize the day's results.
  It will update the previous file list with the new record.

## Environment Setup  
### Specify current comments to extract

```python
TODAY_SUBMISSION_ID = 'n90t4p'
TODAY = '05_10_2021'
```

```python
import numpy as np
import pandas as pd
from datetime import datetime
import datetime, time
import praw as pr
import pickle
from wordcloud import WordCloud
import matplotlib.pyplot as plt
pd.set_option('display.max_colwidth', None)
import sys
sys.path.append('../Packages/')
import diamond_hands as dh
```

```python
raw_comment_dir = '../../Data/RawComments/'
processed_inputs_dir = '../../Data/ProcessedInputs/'
wordcloud_dir = '../../Data/Outputs/Wordcloud/Today/'
processed_comments_dir = '../../Data/Outputs/ProcessedComments/Tomorrow/'

today_comments_all = processed_inputs_dir + 'Today_All_Comment_Summary.bin'
today_comments_tickr = processed_inputs_dir + 'Today_Tickr_Comment_Summary.bin'
tomorrow_comments_all = processed_inputs_dir + 'Tomorrow_All_Comment_Summary.bin'
tomorrow_comments_tickr = processed_inputs_dir + 'Tomorrow_Tickr_Comment_Summary.bin'

logging_file = '../../Data/Log/log.bin'

REDDIT_CLIENT_ID = ""
REDDIT_CLIENT_SECRET = ""
REDDIT_USER_AGENT ="my user agent"
```

```python
# establish reddit connection
reddit = pr.Reddit(
    client_id=REDDIT_CLIENT_ID,
    client_secret=REDDIT_CLIENT_SECRET,
    user_agent=REDDIT_USER_AGENT,
)
```

## Comment Extraction  
Get comments using praw connection. Save results to file.  
**Note: this can take a while**

```python
today_submission = reddit.submission(id=TODAY_SUBMISSION_ID)
```

```python
start_time = datetime.datetime.now()
print('Started: %s' % start_time.strftime("%H:%M:%S.%f"))
print(today_submission.num_comments)
today_submission.comments.replace_more(limit=None)
diff = datetime.datetime.now() - start_time
print("Time taken: %s" % diff)
```

```python
today_comments = today_submission.comments.list()
today_file = raw_comment_dir+'Today/Today_'+TODAY+'.bin'
with open(today_file,'wb') as f:
    pickle.dump(today_comments,f)
```

## Comment Cleaning and Sentiment

```python
today_num_comments, today_summary_all, today_summary_tickr, processed_comments = dh.processComments(today_file)
```

```python
# Load historical data and append new records
with open(today_comments_all,'rb') as f:
    today_all_df = pickle.load(f)
    
with open(today_comments_tickr,'rb') as f:
    today_tickr_df = pickle.load(f)
```

```python
today_summary_all = pd.Series(today_summary_all,index = today_all_df.columns)
today_all_df = today_all_df.append(today_summary_all,ignore_index=True)
```

```python
today_summary_tickr2 = pd.DataFrame(today_summary_tickr,columns = today_tickr_df.columns)
today_tickr_df = today_tickr_df.append(today_summary_tickr2,ignore_index=True)
```

## Visualize Results

```python
dh.plotWordCloud(today_summary_tickr2,filename=wordcloud_dir+TODAY+'.png')
```

## Save Results

```python
# Save results
with open(today_comments_all,'wb') as f:
    pickle.dump(today_all_df,f)
    
with open(today_comments_tickr,'wb') as f:
    pickle.dump(today_tickr_df,f)
```

```python
with open(processed_comments_dir+TODAY+'.bin','wb') as f:
    pickle.dump(processed_comments,f)
```

## Keep track of actions performed

```python
with open(logging_file,'rb') as f:
    model_log = pickle.load(f)

# Update comments processed
model_log['#Today_Processed'] += today_num_comments
model_log['Last_Today'] = TODAY

with open(logging_file,'wb') as f:
    pickle.dump(model_log,f)
```
